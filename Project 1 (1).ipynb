```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setting seaborn style for better aesthetics
sns.set_style('whitegrid')

# Enable inline plotting
%matplotlib inline 
```

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv('depression_data.csv')

df.head()
```

```python
# Check for missing values in each column
print(df.isnull().sum())
```

```python
# Check for duplicate rows
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
# Get information about data types and non-null values
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=False, color='blue', bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(8, 8))
df['Marital Status'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Marital Status Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df[df['Family History of Depression'] == 'Yes']['Smoking Status'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('muted'))
plt.title('Smoking Status Among Individuals with a Family History of Depression')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Age', y='Income', data=df, marker='o', color='green')
plt.title('Income Trend Across Age')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Physical Activity Level'], kde=False, color='purple', bins=10)
plt.title('Distribution of Physical Activity Levels')
plt.xlabel('Physical Activity Level')
plt.ylabel('Number of People')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Education Level', hue='History of Mental Illness', data=df, palette='Set2')
plt.title('Education Level vs. History of Mental Illness')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.legend(title='History of Mental Illness')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['Sleep Patterns'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Distribution of Sleep Patterns')
plt.ylabel('')
plt.show()
```
